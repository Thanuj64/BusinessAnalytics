# ***Understanding Movie Trends in Bollywood***

Data Source: Bollywood is a major film industry with a large cultural impact. Web scraping from IMDb or Bollywood movie databases.


EDA Focus:

1. Analyze box office collections and their relationship with movie genres, actors, and directors.

2. Explore trends in movie releases over the years.

3. Visualize data using bar charts, pie charts, and word clouds (for movie titles or genres).


Pandas Library Usage:

Use libraries like requests and Beautiful Soup for web scraping. String manipulation using Pandas to clean movie titles and other text data.

df.value counts() to analyze frequencies of categories.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample Bollywood Movie Data (Replace with real data from IMDb or other sources)
data = {
    'movie_title': ['Movie A', 'Movie B', 'Movie C', 'Movie D', 'Movie E',
                    'Movie F', 'Movie G', 'Movie H', 'Movie I', 'Movie 1'],
    'genre': ['Action', 'Romance', 'Comedy', 'Drama', 'Action',
              'Comedy', 'Romance', 'Thriller', 'Drama', 'Action'],
    'release_year': [2023, 2022, 2023, 2021, 2022,
                     2023, 2020, 2022, 2021, 2023],
    'box_office_crores': [150, 80, 120, 60, 200,
                          100, 90, 70, 50, 180],
    'lead_actor': ['Actor X', 'Actor Y', 'Actor Z', 'Actor X', 'Actor Y',
                   'Actor Z', 'Actor X', 'Actor Y', 'Actor Z', 'Actor X']
}

# Create DataFrame
bollywood_df = pd.DataFrame(data)

# 1. Basic Data Exploration
print(bollywood_df.head())
print(bollywood_df.info())
```

```python
# 2. Genre Analysis

genre_counts = bollywood_df['genre'].value_counts()

print("\nMovie Count by Genre:\n", genre_counts)
```

```python
# 3. Box Office Collection by Genre

box_office_by_genre = bollywood_df.groupby('genre')['box_office_crores'].sum()

print("\nTotal Box Office Collection by Genre:\n", box_office_by_genre)
```

```python
# 4. Lead Actor Analysis

lead_actor_box_office = bollywood_df.groupby('lead_actor')['box_office_crores'].sum()

print("\nTotal Box Office Collection by Lead Actor:\n", lead_actor_box_office)
```

```python
# 1. Bar Chart of Movie Count by Genre

plt.figure(figsize=(10, 6))

sns.countplot(x='genre', data=bollywood_df, palette='pastel')

plt.title('Number of Movies by Genre')
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45)  # Optional: Rotate x-axis labels if needed
plt.tight_layout()
plt.show()
```

```python
# 2. Pie Chart of Box Office Collection by Genre

plt.figure(figsize=(8, 8))

plt.pie(box_office_by_genre,
        labels=box_office_by_genre.index,
        autopct='%1.1f%%',
        startangle=140,
        colors=sns.color_palette('pastel'))

plt.title('Box Office Collection by Genre')
plt.axis('equal')  # Ensures pie is a circle
plt.show()
```
